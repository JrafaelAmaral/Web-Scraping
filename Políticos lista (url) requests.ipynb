# Raspagem de dados - captura de páginas

Obs.: cada api tem a sua documentação e é necessário ler cada uma de seu interesse antes. Para fazer o scraping é necessário entender como a página foi construída.

Ir em 'Os dados disponíveis' (https://dadosabertos.camara.leg.br/swagger/api.html)

```python
import requests

# função abaixo é mesma coisa que ver o código fonte de uma página.
# devolve HTML
url = 'https://dadosabertos.camara.leg.br/'
resposta = requests.get(url)
print(resposta.text)
```

```python
url = 'https://dadosabertos.camara.leg.br/api/v2/deputados'
parametros = {'formato': 'json'}
resposta = requests.get(url, parametros)
print(resposta.json())
```

```python
# como são 500 deputados no total, é necessário um loop for de 1 à 6

deputados = []
url = 'https://dadosabertos.camara.leg.br/api/v2/deputados'
for pagina in [1, 2, 3, 4, 5, 6]:
    # print(pagina)
    parametros = {'formato': 'json', 'itens': 100, 'pagina': pagina}
    resposta = requests.get(url, parametros)
    for deputado in resposta.json()['dados']:
        deputados.append(deputado['nome'])
print(len(deputados))
```

```python
# como são 500 deputados no total, é necessário um loop for de 1 à 6

deputados = []
url = 'https://dadosabertos.camara.leg.br/api/v2/deputados'
for pagina in [1, 2, 3, 4, 5, 6]:
    parametros = {'formato': 'json', 'itens': 100, 'pagina': pagina}
    resposta = requests.get(url, parametros)
    for deputado in resposta.json()['dados']:
        print(f"{deputado['nome']} ({deputado['siglaPartido']})")
```
